```python
def create_features(df):
    df = df.copy()
    df['Month']     = df.index.month
    df['DayOfWeek'] = df.index.dayofweek
    df['Year']      = df.index.year
    df['Lag_1']     = df['Sales'].shift(1)
    return df

train_feat = create_features(train)
valid_feat = create_features(valid)

train_feat.dropna(subset=['Lag_1'], inplace=True)
valid_feat.dropna(subset=['Lag_1'], inplace=True)

```

```python
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
features = [
    'Month','DayOfWeek','Year','Lag_1',
    'Promo','SchoolHoliday',
    'Open','CompetitionDistance','Promo2'
]
```

```python
X_train = train_feat[features]
y_train = train_feat['Sales']
X_valid = valid_feat[features]
y_valid = valid_feat['Sales']
```

```python
model = xgb.XGBRegressor(
    n_estimators=100, 
    early_stopping_rounds = 130,
)
model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)]
         ,verbose = True
         )
```

```python
print("Best iteration:", model.best_iteration)
print("Best validation RMSE:", model.best_score)  
```

```python
fi = pd.DataFrame(data=model.feature_importances_,
             index=model.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```
